```python
import pyarrow as pa
import pandas as pd
import numpy as np
import os
import pyarrow.parquet as pq
import fastparquet
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ler todos os arquivos csv do diretório e guardar em um objeto
path = '/home/vinicius.vancellote/data/parquet/join/'
all_files = list(filter(lambda x: '.parquet' in x, os.listdir(path)))
all_files_ago = list(filter(lambda x:'G1-2019-08' in x, all_files))
all_files_dez=(list(filter(lambda x:'G1-2019-12' in x, all_files)))
print (all_files_ago)
```

```python
dataset_analise = []
```

```python
for elem in range(0,len(all_files)):
    dfbus = pd.read_parquet('/home/vinicius.vancellote/data/parquet/join/' + all_files[elem],engine='fastparquet')
    df = dfbus.groupby(['data_particao','horario','id_estacao','acumulado_chuva_15_min','acumulado_chuva_1_h','acumulado_chuva_4_h','acumulado_chuva_24_h','acumulado_chuva_96_h'],as_index=False).agg(
        velocidade_inst = pd.NamedAgg(column = 'VELOCITY',aggfunc = 'mean'),
        agitacao_soma = pd.NamedAgg(column = 'DIST',aggfunc='sum'),
        agitacao_media = pd.NamedAgg(column = 'DIST',aggfunc='mean'),
        avgspeed = pd.NamedAgg(column = 'AVGSPEED',aggfunc='mean'),
        qtd_onibus = pd.NamedAgg(column = 'BUSID',aggfunc = 'nunique'),
        qtd_registros = pd.NamedAgg(column = 'BUSID',aggfunc = 'count'),
    )
    dataset_analise.append(df)
    print(all_files[elem])
df_result = pd.concat(dataset_analise,ignore_index=True)
```
